```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier, plot_tree
import seaborn as sns 
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np
import pandas as pd
```

```python
iris = load_iris()
X = iris.data 
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
classifier_nb = GaussianNB()
classifier_tree = DecisionTreeClassifier()
classifier_nb.fit(X_train, y_train)
classifier_tree.fit(X_train, y_train)
```

```python
y_pred_nb = classifier_nb.predict(X_test)
y_pred_tree = classifier_tree.predict(X_test)
```

```python
matrix_nb = confusion_matrix(y_test, y_pred_tree)
print(matrix_nb)
matrix_tree = confusion_matrix(y_test, y_pred_tree)
print(matrix_tree)

score_nb = accuracy_score(y_test,y_pred_nb)
print(score_nb)
score_tree = accuracy_score(y_test,y_pred_nb)
print(score_tree)
```

```python
plt.figure(figsize=(10, 6))
plot_tree(classifier_tree, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.show()
```
